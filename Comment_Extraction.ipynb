```python
import os
import re
import pandas as pd
```

```python
project_name = input("Please enter project name:\n") #You can find the names of the project inside the Downloads Folder of the current repo
```

```python
current_directory = os.getcwd()
```

```python
locate = current_directory+'\\Downloads'+'\\'+project_name+"\\Python_File_List.txt"
```

```python
with open(locate,'r') as f:
    python_files_path_list = f.readlines()
```

```python
multi_line = re.compile('(?:""")(.*?)(?:""")', re.DOTALL) # Regex pattern for multi-line commet
single_line = re.compile('(?:#(.*))') # Regex Pattern for single-line comment
patterns = [multi_line,single_line]
```

```python
comment_list = []
for i in python_files_path_list:
    i = i.replace("\\","/")
    i = i.replace("\n",'')
    with open(i,'r', encoding="utf-8") as g:
        content = g.read()
    
    for index,pattern in enumerate(patterns):
            if re.match(pattern,content):
               print(g.name)
            
            for n in pattern.finditer(content):
                start = n.start()
                lineno = content.count('\n', 0, start) + 1
                #offset = start - content.rfind('\n', 0, start)
                word = n.group(1)
                file_path = g.name
                #print ( lineno,word,offset)
                if(index == 0):
                    flag = 'Multi-Line'
                elif(index == 1): flag = 'Single-Line'
                comment_list.append([file_path, lineno , word, flag])

            
```

```python
dataframe = pd.DataFrame(comment_list,columns=['File_Path','Start_Line','Comment','Flag'])
path_for_comment_csv = current_directory+'\\Downloads'+'\\'+project_name
path_for_comment_csv = path_for_comment_csv.replace("\\",'/')
dataframe.to_csv(os.path.join(path_for_comment_csv,r'comments.csv')) #Creates comments.csv inside every project folder 
```
