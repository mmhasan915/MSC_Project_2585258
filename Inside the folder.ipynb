```python
from bs4 import BeautifulSoup
import requests
```

```python
url = "https://github.com/public-apis/public-apis/tree/master/scripts"
```

```python
results =  requests.get(url)
```

```python
print(results)
```

```python
scrap_data =  BeautifulSoup(results.text,"html.parser")
```

```python
print(scrap_data.prettify())
```

```python
box  = scrap_data.find("div", class_ = "js-details-container Details")
```

```python
print(box.prettify())
```

```python
individual_box_data = box.find_all("a", class_ = "js-navigation-open Link--primary")
```

```python
print(individual_box_data)
```

```python
for links in individual_box_data:
    print (links.get('href'))
```
