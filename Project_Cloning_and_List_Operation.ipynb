```python
import requests
import json
import os
import git
import time
```

```python
start_time = time.time()
# GitHub Rest API Call
response_API = requests.get('https://api.github.com/search/repositories?q=fork:true+is:public+language:Python&per_page=100')
data = response_API.text
parse_data = json.loads(data)
current_directory = os.getcwd()
Downloads = current_directory+'\\Downloads' # Creates Download Folder Inside Directory
os.mkdir(Downloads)
for index, item in enumerate(parse_data['items']):
    print(index,item['name'],item['html_url'] )
    path = os.path.join(Downloads, item['name'])
    os.mkdir(path)
    repo = git.Repo.clone_from(item['html_url'],path) # Clone Project from GitHub Repo to Local Machine
end_time = time.time()
print(end_time-start_time) # Prints the exact time taken by process to complete
```

```python
List_of_Subfolder = os.listdir("Downloads") 
```

```python

for Subfolder in List_of_Subfolder:
    List_of_Files = []
    Subfolders = "Downloads/"+ Subfolder
    Python_File_List_Path = Subfolders + "/Python_File_List.txt" 
    for root, dirs, files in os.walk(Subfolders):
        for file in files:
            if file.endswith(".py"):
                #print(os.path.abspath(os.path.join(root, file)))
                List_of_Files.append(os.path.abspath(os.path.join(root, file))) 
                
    with open(Python_File_List_Path , 'w', encoding="utf-8") as f:
        for items in List_of_Files:
            f.write(items+"\n")     # Creates Python_File_List.txt inside every project folder.

   
```
